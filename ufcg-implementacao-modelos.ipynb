```python
import pandas as pd
import os

# Caminho base para os arquivos
base_path = "E:/Mestrado UFCG/Semestre 2024.2/Dados/Tabelas_0/"

# Lista de tabelas para carregar
tables = ["alunos", "tabela_motivo_evasao", "tabela_dados_ingresso", "tabela_dados_pessoais", "matriculas"]

# Dicionário para armazenar os DataFrames
dataframes = {}

# Carregando as tabelas com o delimitador correto
def load_tables():
    for table in tables:
        file_path = os.path.join(base_path, f"{table}.csv")
        try:
            # Carrega o CSV com o delimitador correto
            dataframes[table] = pd.read_csv(file_path, delimiter=';', encoding='utf-8')
            print(f"Tabela {table} carregada com sucesso!")
        except Exception as e:
            print(f"Erro ao carregar {table}: {e}")

# Executa o carregamento
load_tables()

# Verificando as colunas dos dataframes
for name, df in dataframes.items():
    print(f"\nColunas do DataFrame {name}: {df.columns}")

# Dicionário com as colunas relevantes para cada tabela
colunas_relevantes = {
    "alunos": ["MATRICULA", "GENERO", "IDADE", "ESTADO_CIVIL_ALUNOS", "NACIONALIDADE", "RAZAO_INATIVIDADE",
               "TIPO_ADMISSAO", "POLITICA_AFIRMATIVA"],
    "tabela_motivo_evasao": ["MATRICULA", "RAZAO_DE_INATIVIDADE_DE_ALUNO"],
    "tabela_dados_ingresso": ["MATRICULA", "TIPO_ENSINO_MEDIO", "POLITICA_AFIRMATIVA_ALUNOS"],
    "tabela_dados_pessoais": ["MATRICULA", "IDADE", "GENERO", "LOCAL_NASCIMENTO", "ESTADO"],
    "matriculas": ["MATRICULA", "CODIGO_DISCIPLINA", "CREDITOS", "NOTA", "ESTATUS"]
}

# Dicionário para armazenar os DataFrames com colunas relevantes
dataframes_relevantes = {}

# Selecionando as colunas relevantes
for table, df in dataframes.items():
    try:
        # Verifica se as colunas existem no DataFrame
        colunas = [col for col in colunas_relevantes[table] if col in df.columns]
        if not colunas:
            print(f"Aviso: Nenhuma coluna relevante encontrada na tabela {table}.")
        else:
            dataframes_relevantes[table] = df[colunas]
            print(f"Colunas relevantes da tabela {table} selecionadas com sucesso!")
    except KeyError:
        print(f"Tabela {table} não encontrada na lista de colunas relevantes.")
    except Exception as e:
        print(f"Erro ao selecionar colunas da tabela {table}: {e}")

# Visualizando os DataFrames com colunas relevantes
for table, df in dataframes_relevantes.items():
    print(f"\nTabela: {table}")
    print("Colunas selecionadas:", df.columns.tolist())
    print(df.head())
```

### Escolha de colunas relevantes ###

```python
# Dicionário com as colunas relevantes para cada tabela
colunas_relevantes = {
    "alunos": ["MATRICULA", "GENERO", "IDADE", "ESTADO_CIVIL_ALUNOS", "NACIONALIDADE", "RAZAO_INATIVIDADE",
               "TIPO_ADMISSAO", "POLITICA_AFIRMATIVA"],
    "tabela_motivo_evasao": ["MATRICULA", "RAZAO_DE_INATIVIDADE_DE_ALUNO"],
    "tabela_dados_ingresso": ["MATRICULA", "TIPO_ENSINO_MEDIO", "POLITICA_AFIRMATIVA_ALUNOS"],
    "tabela_dados_pessoais": ["MATRICULA", "IDADE", "GENERO", "LOCAL_NASCIMENTO", "ESTADO"],
    "matriculas": ["MATRICULA", "CODIGO_DISCIPLINA", "CREDITOS", "NOTA", "ESTATUS"]
}

# Garantindo que apenas as colunas relevantes estejam presentes
for table, df in dataframes_relevantes.items():
    try:
        # Seleciona apenas as colunas relevantes
        colunas = [col for col in colunas_relevantes[table] if col in df.columns]
        dataframes_relevantes[table] = df[colunas]
        print(f"Colunas relevantes da tabela {table} garantidas com sucesso!")
    except KeyError:
        print(f"Tabela {table} não encontrada na lista de colunas relevantes.")
    except Exception as e:
        print(f"Erro ao garantir colunas da tabela {table}: {e}")

# Visualizando as tabelas relevantes após a garantia
for table, df in dataframes_relevantes.items():
    print(f"\n=== Tabela: {table} ===")
    print(f"Colunas selecionadas: {df.columns.tolist()}")
    print(f"Total de linhas: {len(df)}")
    print("\nPrimeiras linhas:")
    print(df.head())
```

### Verificar valores Nulos ###

```python
# Verificando valores ausentes em cada tabela
for table, df in dataframes_relevantes.items():
    print(f"\n=== Tabela: {table} ===")
    print("Valores ausentes por coluna:")
    print(df.isnull().sum())  # Conta valores nulos por coluna
    print(f"\nTotal de valores ausentes: {df.isnull().sum().sum()}")  # Total de valores nulos na tabela
```

### Preencher Valores Ausentes com um Valor Padrão ###

```python
# Preenchendo valores ausentes com "Desconhecido"
dataframes_relevantes["tabela_dados_pessoais"]["LOCAL_NASCIMENTO"].fillna("Desconhecido", inplace=True)
dataframes_relevantes["tabela_dados_pessoais"]["ESTADO"].fillna("Desconhecido", inplace=True)

# Verificando novamente os valores ausentes
print(dataframes_relevantes["tabela_dados_pessoais"].isnull().sum())
```

### Fazer o Merge ###

```python
# Merge das tabelas
# Começamos com a tabela principal (alunos)
merged_df = dataframes_relevantes["alunos"]

# Merge com a tabela tabela_motivo_evasao
merged_df = pd.merge(
    merged_df,
    dataframes_relevantes["tabela_motivo_evasao"],
    on="MATRICULA",
    how="left"
)

# Merge com a tabela tabela_dados_ingresso
merged_df = pd.merge(
    merged_df,
    dataframes_relevantes["tabela_dados_ingresso"],
    on="MATRICULA",
    how="left"
)

# Merge com a tabela tabela_dados_pessoais
# Aqui, especificamos quais colunas queremos trazer para evitar duplicação
merged_df = pd.merge(
    merged_df,
    dataframes_relevantes["tabela_dados_pessoais"][["MATRICULA", "LOCAL_NASCIMENTO", "ESTADO"]],
    on="MATRICULA",
    how="left"
)

# Merge com a tabela matriculas
merged_df = pd.merge(
    merged_df,
    dataframes_relevantes["matriculas"],
    on="MATRICULA",
    how="left"
)

# Verificando colunas duplicadas
print("Colunas após o merge:")
print(merged_df.columns.tolist())

# Removendo colunas duplicadas (se houver)
# Aqui, garantimos que apenas as colunas originais sejam mantidas
merged_df = merged_df.loc[:, ~merged_df.columns.duplicated()]

# Renomeando colunas (se necessário)
# Se houver colunas renomeadas incorretamente, podemos corrigir
merged_df.rename(columns={
    "GENERO_x": "GENERO",
    "IDADE_x": "IDADE"
}, inplace=True)

# Visualizando o DataFrame final
print("\n=== DataFrame após o merge e correção de colunas ===")
print(f"Total de linhas: {len(merged_df)}")
print(f"Colunas: {merged_df.columns.tolist()}")
print("\nPrimeiras linhas:")
print(merged_df.head())
```

###  Remover Linhas com NaN ###

```python
merged_df.dropna(inplace=True)
```

```python
# Verificando valores ausentes por coluna
print("Valores ausentes por coluna:")
print(merged_df.isnull().sum())

# Verificando o total de valores ausentes
print(f"\nTotal de valores ausentes: {merged_df.isnull().sum().sum()}")
```

### Distribuição de Variáveis Numéricas ### 

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configuração do estilo dos gráficos
sns.set(style="whitegrid")

# Histograma para IDADE
plt.figure(figsize=(8, 6))
sns.histplot(merged_df["IDADE"], kde=True, color="blue")
plt.title("Distribuição da Idade dos Alunos")
plt.xlabel("Idade")
plt.ylabel("Frequência")
plt.show()

# Histograma para NOTA
plt.figure(figsize=(8, 6))
sns.histplot(merged_df["NOTA"], kde=True, color="green")
plt.title("Distribuição das Notas dos Alunos")
plt.xlabel("Nota")
plt.ylabel("Frequência")
plt.show()
```

###  Relação entre Variáveis Numéricas ###

```python
# Gráfico de dispersão: IDADE vs NOTA
plt.figure(figsize=(8, 6))
sns.scatterplot(x="IDADE", y="NOTA", data=merged_df, hue="GENERO", palette="viridis")
plt.title("Relação entre Idade e Nota")
plt.xlabel("Idade")
plt.ylabel("Nota")
plt.legend(title="Gênero")
plt.show()
```

### Contagem de Variáveis Categóricas ###

```python
# Contagem de gênero
plt.figure(figsize=(8, 6))
sns.countplot(x="GENERO", data=merged_df, palette="Set2")
plt.title("Distribuição de Gênero")
plt.xlabel("Gênero")
plt.ylabel("Contagem")
plt.show()

# Contagem de estado civil
plt.figure(figsize=(8, 6))
sns.countplot(x="ESTADO_CIVIL_ALUNOS", data=merged_df, palette="Set3")
plt.title("Distribuição de Estado Civil")
plt.xlabel("Estado Civil")
plt.ylabel("Contagem")
plt.xticks(rotation=45)
plt.show()
```

### Relação entre Variáveis Categóricas e Numéricas ###

```python
# Box plot: NOTA por GÊNERO
plt.figure(figsize=(8, 6))
sns.boxplot(x="GENERO", y="NOTA", data=merged_df, palette="pastel")
plt.title("Distribuição de Notas por Gênero")
plt.xlabel("Gênero")
plt.ylabel("Nota")
plt.show()

# Box plot: IDADE por ESTADO_CIVIL_ALUNOS
plt.figure(figsize=(10, 6))
sns.boxplot(x="ESTADO_CIVIL_ALUNOS", y="IDADE", data=merged_df, palette="coolwarm")
plt.title("Distribuição de Idade por Estado Civil")
plt.xlabel("Estado Civil")
plt.ylabel("Idade")
plt.xticks(rotation=45)
plt.show()
```

### Correlação entre Variáveis Numéricas ###

```python
# Calculando a matriz de correlação
correlacao = merged_df[["IDADE", "CREDITOS", "NOTA"]].corr()

# Heatmap de correlação
plt.figure(figsize=(8, 6))
sns.heatmap(correlacao, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Mapa de Calor de Correlação")
plt.show()
```

### Distribuição de Variáveis por Categoria ###

```python
# Violin plot: NOTA por TIPO_ADMISSAO
plt.figure(figsize=(10, 6))
sns.violinplot(x="TIPO_ADMISSAO", y="NOTA", data=merged_df, palette="Set2")
plt.title("Distribuição de Notas por Tipo de Admissão")
plt.xlabel("Tipo de Admissão")
plt.ylabel("Nota")
plt.xticks(rotation=45)
plt.show()
```

### Contagem de Razões de Inatividade ###

```python
# Contagem de razões de inatividade
plt.figure(figsize=(10, 6))
sns.countplot(y="RAZAO_DE_INATIVIDADE_DE_ALUNO", data=merged_df, palette="Set3", order=merged_df["RAZAO_DE_INATIVIDADE_DE_ALUNO"].value_counts().index)
plt.title("Distribuição de Razões de Inatividade")
plt.xlabel("Contagem")
plt.ylabel("Razão de Inatividade")
plt.show()
```
