```python
import pandas as pd
import os

# Caminho base para os arquivos
base_path = "E:/Mestrado UFCG/Semestre 2024.2/Dados/Tabelas_0/"

# Lista de tabelas para carregar
tables = ["alunos", "tabela_motivo_evasao", "tabela_dados_ingresso", "tabela_dados_pessoais", "matriculas"]

# Dicionário para armazenar os DataFrames
dataframes = {}

# Carregando as tabelas com o delimitador correto
def load_tables():
    for table in tables:
        file_path = os.path.join(base_path, f"{table}.csv")
        try:
            # Carrega o CSV com o delimitador correto
            dataframes[table] = pd.read_csv(file_path, delimiter=';', encoding='utf-8')
            print(f"Tabela {table} carregada com sucesso!")
        except Exception as e:
            print(f"Erro ao carregar {table}: {e}")

# Executa o carregamento
load_tables()

# Verificando as colunas dos dataframes
for name, df in dataframes.items():
    print(f"\nColunas do DataFrame {name}: {df.columns}")

# Dicionário com as colunas relevantes para cada tabela
colunas_relevantes = {
    "alunos": ["MATRICULA", "GENERO", "IDADE", "ESTADO_CIVIL_ALUNOS", "NACIONALIDADE", "RAZAO_INATIVIDADE",
               "TIPO_ADMISSAO", "POLITICA_AFIRMATIVA"],
    "tabela_motivo_evasao": ["MATRICULA", "RAZAO_DE_INATIVIDADE_DE_ALUNO"],
    "tabela_dados_ingresso": ["MATRICULA", "TIPO_ENSINO_MEDIO", "POLITICA_AFIRMATIVA_ALUNOS"],
    "tabela_dados_pessoais": ["MATRICULA", "IDADE", "GENERO", "LOCAL_NASCIMENTO", "ESTADO"],
    "matriculas": ["MATRICULA", "CODIGO_DISCIPLINA", "CREDITOS", "NOTA", "ESTATUS"]
}

# Dicionário para armazenar os DataFrames com colunas relevantes
dataframes_relevantes = {}

# Selecionando as colunas relevantes
for table, df in dataframes.items():
    try:
        # Verifica se as colunas existem no DataFrame
        colunas = [col for col in colunas_relevantes[table] if col in df.columns]
        if not colunas:
            print(f"Aviso: Nenhuma coluna relevante encontrada na tabela {table}.")
        else:
            dataframes_relevantes[table] = df[colunas]
            print(f"Colunas relevantes da tabela {table} selecionadas com sucesso!")
    except KeyError:
        print(f"Tabela {table} não encontrada na lista de colunas relevantes.")
    except Exception as e:
        print(f"Erro ao selecionar colunas da tabela {table}: {e}")

# Visualizando os DataFrames com colunas relevantes
for table, df in dataframes_relevantes.items():
    print(f"\nTabela: {table}")
    print("Colunas selecionadas:", df.columns.tolist())
    print(df.head())
```

### Fazer o Merge ###

```python
# Merge das tabelas
# Começamos com a tabela principal (alunos)
merged_df = dataframes_relevantes["alunos"]

# Merge com a tabela tabela_motivo_evasao
merged_df = pd.merge(
    merged_df,
    dataframes_relevantes["tabela_motivo_evasao"],
    on="MATRICULA",
    how="left"
)

# Merge com a tabela tabela_dados_ingresso
merged_df = pd.merge(
    merged_df,
    dataframes_relevantes["tabela_dados_ingresso"],
    on="MATRICULA",
    how="left"
)

# Merge com a tabela tabela_dados_pessoais
# Aqui, especificamos quais colunas queremos trazer para evitar duplicação
merged_df = pd.merge(
    merged_df,
    dataframes_relevantes["tabela_dados_pessoais"][["MATRICULA", "LOCAL_NASCIMENTO", "ESTADO"]],
    on="MATRICULA",
    how="left"
)

# Merge com a tabela matriculas
merged_df = pd.merge(
    merged_df,
    dataframes_relevantes["matriculas"],
    on="MATRICULA",
    how="left"
)

# Verificando colunas duplicadas
print("Colunas após o merge:")
print(merged_df.columns.tolist())

# Removendo colunas duplicadas (se houver)
# Aqui, garantimos que apenas as colunas originais sejam mantidas
merged_df = merged_df.loc[:, ~merged_df.columns.duplicated()]

# Renomeando colunas (se necessário)
# Se houver colunas renomeadas incorretamente, podemos corrigir
merged_df.rename(columns={
    "GENERO_x": "GENERO",
    "IDADE_x": "IDADE"
}, inplace=True)

# Visualizando o DataFrame final
print("\n=== DataFrame após o merge e correção de colunas ===")
print(f"Total de linhas: {len(merged_df)}")
print(f"Colunas: {merged_df.columns.tolist()}")
print("\nPrimeiras linhas:")
print(merged_df.head())
```

```python
# Exemplo de separação dos dados em subgrupos, por exemplo, alunos com alta média (nota >= 7)
grupo_alunos_bom_desempenho = X[X['nota_media'] >= 7]  # Filtrando para alunos com boa nota
grupo_alunos_bom_desempenho_target = y[X['nota_media'] >= 7]

grupo_alunos_baixo_desempenho = X[X['nota_media'] < 7]  # Filtrando para alunos com baixo desempenho
grupo_alunos_baixo_desempenho_target = y[X['nota_media'] < 7]

# Dividindo em treino e teste para o subgrupo de bom desempenho
X_train_bom, X_test_bom, y_train_bom, y_test_bom = train_test_split(grupo_alunos_bom_desempenho, grupo_alunos_bom_desempenho_target, test_size=0.3, random_state=42)

# Treinando o modelo para o subgrupo de bom desempenho
modelo_bom_desempenho = RandomForestClassifier(n_estimators=100, random_state=42)
modelo_bom_desempenho.fit(X_train_bom, y_train_bom)

# Previsões para o subgrupo de bom desempenho
y_pred_bom = modelo_bom_desempenho.predict(X_test_bom)

# Avaliando o desempenho para o subgrupo de bom desempenho
print("Modelo Específico - Bom Desempenho:")
print(classification_report(y_test_bom, y_pred_bom))
print("Matriz de Confusão:")
print(confusion_matrix(y_test_bom, y_pred_bom))
print("Kappa:", cohen_kappa_score(y_test_bom, y_pred_bom))


# Dividindo em treino e teste para o subgrupo de baixo desempenho
X_train_baixo, X_test_baixo, y_train_baixo, y_test_baixo = train_test_split(grupo_alunos_baixo_desempenho, grupo_alunos_baixo_desempenho_target, test_size=0.3, random_state=42)

# Treinando o modelo para o subgrupo de baixo desempenho
modelo_baixo_desempenho = RandomForestClassifier(n_estimators=100, random_state=42)
modelo_baixo_desempenho.fit(X_train_baixo, y_train_baixo)

# Previsões para o subgrupo de baixo desempenho
y_pred_baixo = modelo_baixo_desempenho.predict(X_test_baixo)

# Avaliando o desempenho para o subgrupo de baixo desempenho
print("Modelo Específico - Baixo Desempenho:")
print(classification_report(y_test_baixo, y_pred_baixo))
print("Matriz de Confusão:")
print(confusion_matrix(y_test_baixo, y_pred_baixo))
print("Kappa:", cohen_kappa_score(y_test_baixo, y_pred_baixo))
```
